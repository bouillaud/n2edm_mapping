```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.size"]=12
# plt.rcParams["lines.markersize"]=8
# plt.rcParams["figure.figsize"]=[12, 8]
# plt.rcParams["figure.dpi"]=200
plt.rcParams["axes.labelsize"]=15
```

```python
G10s = np.load("../data/G10s_runstart1095_runstop1150_lmax1.npy")
G10s3 = np.load("../data/G10s_runstart1095_runstop1150_lmax3.npy")
```

```python
G10s
```

```python
runs = range(1095, 1150)

plt.plot(runs, G10s, "-o")
plt.grid()

plt.ylim(-1, 0)
plt.yticks(np.arange(-1, 0.1, 0.1))
plt.xlim(1095, 1150)

plt.xlabel("Run number")
plt.ylabel(r"$G_{10}$ [pT/cm]")
# plt.savefig("../plots/G10s_runstart1095_runstop1150.pdf")
```

```python
sum(G10s[0:5])
```

```python
dl = 3
G10s_smooth = []
for i in range(dl, 55-dl):
    G10s_smooth.append(sum(G10s[i-dl:i+dl])/(2*dl))
```

```python
plt.plot(np.arange(1095+dl, 1150-dl), G10s_smooth, "-o")
plt.grid()

plt.ylim(-1, 0)
plt.yticks(np.arange(-1, 0.1, 0.1))
plt.xlim(1095, 1150)

plt.xlabel("Run number")
plt.ylabel(r"$G_{10}$ [pT/cm]")
# plt.savefig("../plots/G10s_runstart1095_runstop1150_avg5runs.pdf")
```
